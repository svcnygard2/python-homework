```python
#import correct libraries
import csv
from pathlib import Path
%matplotlib inline
import numpy as np

#declare variables that will be used
date=""
pnl=0
count_month=0
total_pnl=0
average_change=0
minimum=0
minimum_key=""
maximum=0
maximum_key=""
pnl_change=0
pnl_prior=0
pnl_change_sum=0
month_one = 0


#pull .csv file 
new_path=Path("budget_data.csv")

#open .csv file and start reading the data on the "next" line
with open(new_path) as csv_file:
    next(csv_file)
    csv_reader = csv.reader(csv_file, delimiter=',')

    #declared and named both columns "date" and "pnl"
    for data in csv_reader:
        date = data[0]
        pnl = float(data[1])
        
        #set a variable for the first month
        if month_one == 0:
            month_one = pnl

        #created equations to calculate the parameters
        #asked for and counted the months
        total_pnl += pnl
        pnl_change = pnl - pnl_prior
        pnl_change_sum += pnl_change
        pnl_change_sum_m1 = pnl_change_sum - month_one       
        pnl_prior = pnl
        count_month += 1
        count_pnl_change = count_month - 1

        #created a loop to look through the numbers
        #and decide the max and min values and dates
        if minimum == 0:
            minimum = pnl_change
            minimum_key = date
        elif pnl_change < minimum:
            minimum = pnl_change
            minimum_key = date
        elif pnl_change > maximum:
            maximum = pnl_change
            maximum_key = date

#calculated average_change result            
average_change = round((pnl_change_sum_m1 / count_pnl_change), 2)

#Print final results
print("Financial Analysis")
print("----------------------------")
print(f"Total Months: {count_month}")
print(f"Total: ${total_pnl}")
print(f"Average Change: ${average_change}")
print(f"Greatest Increase in Profits: {maximum_key} (${maximum})")
print(f"Greatest Decrease in Profits: {minimum_key} (${minimum})")

#Set output file name
output_path = 'output.txt'

#Open the output path and print it to a .txt file
with open(output_path, 'w') as file:
    
    file.write("Financial Analysis\n")
    file.write("----------------------------\n")
    file.write(f"Total Months: {count_month}\n")
    file.write(f"Total: ${total_pnl}\n")
    file.write(f"Average Change: ${average_change}\n")
    file.write(f"Greatest Increase in Profits: {maximum_key} ${maximum}\n")
    file.write(f"Greatest Decrease in Profits: {minimum_key} ${minimum}")
```
